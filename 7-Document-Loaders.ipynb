### Importing Packages

```python
from dotenv import load_dotenv
from langchain_community.document_loaders import TextLoader, DirectoryLoader, PyPDFLoader, WebBaseLoader, CSVLoader

load_dotenv()
```

### TextLoader

```python
loader = TextLoader("data/cricket.txt", encoding="utf-8")

docs = loader.load()

print(docs)
```

### PyPDFLoader

```python
loader = PyPDFLoader("data/dl-curriculum.pdf")
docs = loader.load()

print(len(docs))
print(docs[0].page_content)
```

### Directory Loader

```python
loader = DirectoryLoader(
    path="data/pdfDocs", 
    glob="*.pdf", 
    loader_cls=PyPDFLoader
)
```

```python
docs = loader.load()

for doc in docs:
    print(doc.metadata)
```

```python
docs = loader.lazy_load()

for doc in docs:
    print(doc.metadata)
```

### WebBaseLoader

### CSVLoader

```python
loader = CSVLoader("data/Social_Network_Ads.csv")

docs = loader.load()

for doc in docs:
    print(doc.page_content)
```
